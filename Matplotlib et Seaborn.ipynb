## Style orienté objet et méthode pyplot

Il existe deux facons de tracer un graphe avec matplotlib : le style orienté objet et la méthode pyplot

### le sytle orienté objet 

```python
# générer 100 valeurs dans l'intervalle [0,2]

x = np.linspace(0,2,100)

# Représenter un repère de taille 10 pouces x 5 pouces 

fig,ax = plt.subplots(figsize = (10,5), constrained_layout = True)

# f(x) = x

ax.plot(x,x,label = 'linear')

# f(x) = x ** 2 

ax.plot(x,x**2,label='quadratic')

# f(x) = x ** 3

ax.plot(x,x**3,label='cubic')

# nommer l'axe des abscisses 

ax.set_xlabel('Axe 1')

# nommer l'axe des ordonnées 

ax.set_ylabel('Axe 2')

# donner un titre au graphe

ax.set_title('Simple graphe')

# Ajout de la légende 

ax.legend()
```

### La méthode pyplot

```python
# générer 100 valeurs dans l'intervalle [0,2]

x = np.linspace(0,2,100)

# Représenter un repère de taille 10 pouces x 5 pouces 

fig,ax = plt.subplots(figsize = (10,5), constrained_layout = True)

# f(x) = x

ax.plot(x,x,label = 'linear')

# f(x) = x ** 2 

ax.plot(x,x**2,label='quadratic')

# f(x) = x ** 3

ax.plot(x,x**3,label='cubic')


# nommer l'axe des abscisses 

plt.xlabel('Axe 1')

# nommer l'axe des ordonnées 

plt.ylabel('Axe 2')

# donner un titre au graphe

plt.title('Simple graphe')

# Ajout de la légende 

plt.legend()
```

# Méthode subplot_mosaic

## Exemple 1

```python
fig,axd = plt.subplot_mosaic([['upleft','upright'],['lowleft','lowright']],constrained_layout = True)
axd['upleft'].set_title('upleft')
axd['upright'].set_title('upright')
axd['lowleft'].set_title('lowleft')
axd['lowright'].set_title('lowright')
```

## Exemple 2

```python
fig,axd = plt.subplot_mosaic([['upleft','right'],['lowleft','right']], constrained_layout=True)
axd['upleft'].set_title('upleft')
axd['lowleft'].set_title('lowleft')
axd['right'].set_title('right')
```

# Graphes pour variables catégorielles

## Diagramme en bandes ou en bâtons

```python
names = ['group_a', 'group_b', 'group_c']

values = [1,10,100]

plt.figure(figsize = (12,5))

# mode bar

plt.subplot(131)

plt.bar(names,values)

# mode scatter

plt.subplot(132)

plt.scatter(names,values)

# mode plot

plt.subplot(133)

plt.plot(names,values)

# Ajouter un titre au graphe

plt.suptitle('Tracé pour variables catégorielles')

# Afficher le graphe

plt.show()
```
